# Data Cleaning

```python
#Replacing '?' with 'NAN'

diabetes = diabetes.replace('?', np.nan)
```

```python
diabetes.head()
```

```python
#Removing pareanthesis around age column

diabetes['age'] = diabetes['age'].str.replace('[','').str.replace(')','')
```

```python
#Checking for null values
missing_val = diabetes.isnull().sum().sort_values(ascending=False)
missing_val
```

```python
#Perccentage of missing values

missing_pct = missing_val[missing_val != 0]/len(diabetes) * 100
missing_pct
```

```python
#Removing columns with more missing values

diabetes_cleaned = diabetes.drop(['weight', 'medical_specialty', 'payer_code'], axis =1)
```

```python
#Columns with low missing value is replaced with mode and mean

diabetes_cleaned['race'] = diabetes_cleaned['race'].fillna(diabetes_cleaned['race'].mode()[0])

diabetes_cleaned['diag_1'] = diabetes_cleaned['diag_1'].fillna(diabetes_cleaned['diag_1'].mode()[0])
diabetes_cleaned['diag_2'] = diabetes_cleaned['diag_2'].fillna(diabetes_cleaned['diag_2'].mode()[0])
diabetes_cleaned['diag_3'] = diabetes_cleaned['diag_3'].fillna(diabetes_cleaned['diag_3'].mode()[0])
```
